```python
import numpy as numpy
import pandas as pd
# from ipynb.fs.full.data_preprocess import c_fraud_df
c_fraud_df = pd.read_csv("corrected_names.csv")
from sklearn.preprocessing import OneHotEncoder
```

```python
c_fraud_df.tail(10)
```

```python
c_fraud_df.drop('Unnamed: 0',axis = 1, inplace = True)
```

```python
c_fraud_df
```

```python
array = c_fraud_df["Call Type"].values.reshape(-1,1)
array
```

```python
encoder = OneHotEncoder()
encoded_values = encoder.fit_transform(array).toarray()
encoded_values
```

```python
encoded_df = pd.DataFrame(encoded_values, columns=encoder.get_feature_names_out(['Call Type']))
encoded_df = encoded_df.astype(int)
encoded_df
```

```python
final_df = pd.concat([c_fraud_df,encoded_df], axis=1).drop('Call Type', axis = 1)
final_df.head()
```
